## Imports

```python
from source_code.utils import load_dataset, preprocess, show_sample, evaluate_error, save_model, load_model
from source_code.models import kSVD, TWI_kSVD
```

## Loading data

```python
dataset_name = 'BME'
train_dataset, test_dataset = load_dataset(dataset_name)

train_matrix, train_labels = preprocess(train_dataset, serie_length=128)
```

## Training models

```python
for sparsity in [2, 5, 10]:
    model = kSVD(max_iter=10, verbose=True).fit(train_matrix, sparsity=sparsity)
    save_model(model, f'kSVD_{sparsity}_atoms')
    twi_model = TWI_kSVD(max_iter=10, verbose=True).fit(train_matrix, sparsity=sparsity)
    save_model(twi_model, f'TWI_kSVD_{sparsity}_atoms')
```
